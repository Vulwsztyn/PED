```python
import matplotlib.pyplot as plt
import pickle

comments_circa_397 = pickle.load(open( "../../pickle/comments_circa_397.pkl", "rb" ))
comments_circa_1044 = pickle.load(open( "../../pickle/comments_circa_1044.pkl", "rb" ))
comments_circa_97998 = pickle.load(open( "../../pickle/comments_circa_97998.pkl", "rb" ))
comments_circa_97998
```

```python
times_circa_397 = {}
for id in comments_circa_397:
    times_circa_397[id] = []
    for comment in comments_circa_397[id]:
        times_circa_397[id].append(comment.created_utc)
        
time_circa_1044 = {}
for id in comments_circa_1044:
    time_circa_1044[id] = []
    for comment in comments_circa_1044[id]:
        time_circa_1044[id].append(comment.created_utc)
        
time_circa_97998 = {}
for id in comments_circa_97998:
    time_circa_97998[id] = []
    for comment in comments_circa_97998[id]:
        time_circa_97998[id].append(comment.created_utc)
```

```python
for id in times_circa_397:
    plt.hist(times_circa_397[id][1:], bins=50)
    plt.ylabel('Count')
    plt.yscale('log')
    plt.xlabel('Data')

    plt.show()
    
```

```python
for id in time_circa_1044:
    plt.hist(time_circa_1044[id][1:], bins=50)
    plt.ylabel('Count')
    plt.yscale('log')
    plt.xlabel('Data')

    plt.show()
    
```
